### Q5) GroupBy Question
##### 1. Subset the NHL_Goalies DataFrame where Injuries is not known <font color='red'>(Injuries=NaN)</font>; 
##### 2. Use a Merge with the GAA_  DF to produce a NEW DF (*called mergedDF*) 
###### *** USE A JOIN THAT PRESERVES ORDER AND USES THE INTERSECTION OF KEYS***
###### *** USE BOTH THE LEFT AND RIGHT INDEXES AS JOIN KEYS***
##### 3. From the mergedDF keep FirstName/LastName/Team/Cntry/ SV% /GA/GAA/ TOI (note one copy for any duplicate column)
##### 4. GroupBy country on mergedDF and aggregate The Means, Mins, and Maximum of the kept columns
#### 5. Create a dataframe of the <font color='red'>TOI </font>groupby object and write it to a comma seperated value file called ('TOI_2017.csv')

## Notes:

```python
#Answer 5 here: please comments
#load the libraries using the normal convention
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pprint
##prevent plotting errors
%matplotlib inline

#Load our excel files of interest into pandas
NHL_Goalies = pd.read_excel('NHLGoalies2016_2017.xls',na_values='',sheet_name=0) 
GAA= pd.read_excel('NHLGoalies2016_2017.xls',na_values='',sheet_name='5vs5')

#filter by Null values of injury row
injuryNA = NHL_Goalies[NHL_Goalies['Injuries'].isnull()]
injuryNA
#merge using THE INTERSECTION OF KEYS
mergeDF = pd.merge(injuryNA,GAA,left_index=True,right_index=True)

#print(mergeDF.iloc[0])
mergeDF

#print(list(mergeDF.columns))

mergeDF=mergeDF[['Last Name_x', 'First Name_x', 'Team_x', 'Cntry', 'SV%_x','GA_x','GAA_x', 'TOI']]
mergeDF.columns = ['Last Name', 'First Name', 'Team', 'Cntry', 'SV','GA','GAA', 'TOI']
#mergeDF['Last Name_x']
#mergeDF['First Name_x']
#'Team_x'
###RANDOM NOTES
#filtered_df = filtered_df[(df['C']=='True')
#nbs = df['name'].str.extract('^(N/A|NA|na|n/a)')
#nms=df[(df['name'] != nbs)]


## using .groupby function to see mean of columns by conttry
mergeDF.groupby('Cntry').SV.mean()
mergeDF.groupby('Cntry').GA.mean()
mergeDF.groupby('Cntry').GAA.mean()
mergeDF.groupby('Cntry').TOI.mean()


# max number of values by conttry
mergeDF.groupby('Cntry').SV.max()
mergeDF.groupby('Cntry').GA.max()
mergeDF.groupby('Cntry').GAA.max()
mergeDF.groupby('Cntry').TOI.max()

# min number of values by conttry
mergeDF.groupby('Cntry').SV.min()
mergeDF.groupby('Cntry').GA.min()
mergeDF.groupby('Cntry').GAA.min()
mergeDF.groupby('Cntry').TOI.min()

#Aggregate findings
mergeDF.groupby('Cntry').SV.agg(['mean', 'min', 'max'])
mergeDF.groupby('Cntry').GA.agg(['mean', 'min', 'max'])
mergeDF.groupby('Cntry').GAA.agg(['mean', 'min', 'max'])
mergeDF.groupby('Cntry').TOI.agg(['mean', 'min', 'max'])

allTOI=mergeDF.groupby('Cntry').TOI.agg(['mean', 'min', 'max'])

allTOI.to_csv(r'/Users/MOOSE/Desktop/TOI_2017.csv', sep=',')
```

```python
####################Best answer######

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
##prevent plotting errors
%matplotlib inline

#Load our excel files of interest into pandas
NHL_Goalies = pd.read_excel('NHLGoalies2016_2017.xls',na_values='',sheet_name=0) 
GAA= pd.read_excel('NHLGoalies2016_2017.xls',na_values='',sheet_name='5vs5')


def get_subset_by_IQR(df,column):
    #Check if interger
    np.issubdtype(df[column].dtype, np.number)
    if False:
        print('Column is not numercal')
    else:
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        #Just switched q1 and q3 to get outliers
        notiqr = (df[column] >= q3) | (df[column] <= q1)
    return df.loc[notiqr]

result= get_subset_by_IQR(NHL_Goalies, 'Wt')
result
```
